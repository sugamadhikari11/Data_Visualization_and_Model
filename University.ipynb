```python
import pandas as pd
import numpy as np
import timeit
```

```python
# Load the university ranking data
data = pd.read_csv('/content/University_Dataset.csv')
data = pd.DataFrame(data)
```

```python
data = data.drop(["Overall score", "Index"], axis=1)
```

```python
# Remove the percentage sign and convert numeric columns to numeric values
numeric_columns = data.columns.difference(['University name'])
data[numeric_columns] = data[numeric_columns].replace('%', '', regex=True).apply(pd.to_numeric)

# Calculate the mean of each numeric column
mean_values = data[numeric_columns].mean()

# Fill 0% values with mean respective to their own column
data[numeric_columns] = data[numeric_columns].mask(data[numeric_columns] == 0, other=mean_values, axis=1)

# Add the percentage sign back
data[numeric_columns] = data[numeric_columns].astype(int).astype(str) + '%'

# Print the updated DataFrame
data
```

```python
# Modify university names from 'University of [Name]' to '[Name] University'
data['University name'] = data['University name'].str.replace(r'University of (\w+)', r'\1 University')
```

```python
# Define the criteria
criteria = ['Student satisfaction', 'Graduate prospects', 'Research quality', 'Research intensity', 'Entry standards']

# Print a menu for the user
print("Please select the order of criteria by entering the corresponding numbers, separated by commas:")
for i, criterion in enumerate(criteria, 1):
    print(f"{i}. {criterion}")

# Get the user's selection
user_selection = input()

# Convert the user's selection to a list of integers
user_order = [int(num) for num in user_selection.split(',')]

# Assert that the user's selection is within the valid range
assert all(1 <= num <= len(criteria) for num in user_order), "Invalid selection"

# Create the order of criteria for calculations based on the user's selection
criteria_order_calc = [criteria[i-1] + '_calc' for i in user_order]

# Assert that all criteria are included in the calculation order
assert all(criterion + '_calc' in criteria_order_calc for criterion in criteria), "Not all criteria included in calculation order"
```

```python
# Pairwise Comparisons
comparison_matrix = university_data[criteria_order_calc].values
```

```python
print(comparison_matrix)
```

```python
def calculate_criterion_weights():
    # Calculate the weights of the criteria based on pairwise comparisons
    row_sums = comparison_matrix.sum(axis=1)
    col_sums = comparison_matrix.sum(axis=0)

    criterion_weights = []
    for i in range(len(col_sums)):
        criterion_weight = float(row_sums[i]) / float(col_sums[i])
        criterion_weights.append(criterion_weight)

    # Normalize the weights
    sum_weights = sum(criterion_weights)
    criterion_weights = [weight / sum_weights for weight in criterion_weights]

    # Assert that the sum of weights is 1 (or very close to 1 due to floating point precision)
    assert abs(sum(criterion_weights) - 1) < 1e-6, "Sum of weights is not 1"

    return criterion_weights

def calculate_university_scores(criterion_weights):
    # Perform weighted sum calculations for each university based on criterion weights
    university_scores = []
    for i in range(len(university_data)):
        university_score_elements = []
        for j in range(len(criterion_weights)):
            print(f"university_data.iloc[{i}, {j+1}]: {university_data.iloc[i, j+1]}")
            print(f"criterion_weights[{j}]: {criterion_weights[j]}")
            element = float(university_data.iloc[i, j+1].rstrip('%')) / 100.0 * criterion_weights[j]
            university_score_elements.append(element)
        university_score = sum(university_score_elements)
        university_scores.append(university_score)

    # Assert that all scores are between 0 and 1
    assert all(0 <= score <= 1 for score in university_scores), "Some scores are not between 0 and 1"

    # Update the university_data DataFrame with the overall scores
    university_data['Overall score'] = university_scores
```

```python

criterion_weights = calculate_criterion_weights()
print(criterion_weights)

calculate_university_scores(criterion_weights)
print(university_data)
```

```python
# Merge Sort
def merge_sort(univ_list, criteria):
    # Assert that the criteria is in the university data
    assert all(criteria in univ for univ in univ_list), "Criteria not in university data"

    if len(univ_list) <= 1:
        return univ_list

    mid = len(univ_list) // 2
    left_half = univ_list[:mid]
    right_half = univ_list[mid:]

    # Assert that the list was split correctly
    assert len(left_half) + len(right_half) == len(univ_list), "Failed to split list correctly"

    return merge(merge_sort(left_half, criteria), merge_sort(right_half, criteria), criteria)

def merge(left, right, criteria):
    merged = []
    left_index = 0
    right_index = 0

    # Merge smaller elements first
    while left_index < len(left) and right_index < len(right):
        if left[left_index][criteria] > right[right_index][criteria]:
            merged.append(left[left_index])
            left_index += 1
        elif left[left_index][criteria] < right[right_index][criteria]:
            merged.append(right[right_index])
            right_index += 1
        else:  # If the criteria values are equal
            if left[left_index]['University name'] < right[right_index]['University name']:
                merged.append(left[left_index])
                left_index += 1
            else:
                merged.append(right[right_index])
                right_index += 1

    # If there are remaining elements in left or right half, append them to the result
    while left_index < len(left):
        merged.append(left[left_index])
        left_index += 1

    while right_index < len(right):
        merged.append(right[right_index])
        right_index += 1

    return merged


# Hash-based Search
def hash_search(univ_list, target):
    univ_dict = {univ['University name']: univ for univ in univ_list}

    # Assert that the target university is in the data
    assert target in univ_dict, "Target university not found"

    return univ_dict.get(target, None)

```

```python
# User Interaction for University Search
while True:
    print("Welcome to University Ranking")
    print("Please choose an option:")
    print("1. Display the ranking of all universities")
    print("2. Search for a specific university")
    print("3. Exit")
    option = input("Enter the selection between 1 or 2 or 3: ")

    # Assert that the user's input is a valid option
    assert option in ['1', '2', '3'], "Invalid option"

    if option == '1':
        print("Please enter the criteria you want to sort by:")
        criteria_input = input()
        # Convert the input to the desired format
        criteria_input = criteria_input.lower().capitalize()
        print("Searching for university...")



        start_time = timeit.default_timer()
        # Sort the universities
        sorted_universities = merge_sort(university_dict, criteria_input)

        # Record the ending time
        end_time = timeit.default_timer()

        # Calculate the execution time in seconds
        execution_time = end_time - start_time


        # Print or store the execution time
        print(f"Execution time: {execution_time} seconds")

        # Print the ranking of all universities
        for i, univ in enumerate(sorted_universities):
            if criteria_input == 'Graduate prospects':
                print(f"Rank {i+1}: {univ['University name']} - {criteria_input}: {univ[criteria_input]}%")
            else:
                print(f"Rank {i+1}: {univ['University name']} - {criteria_input}: {univ[criteria_input]}")
    elif option == '2':
        print("Please enter the name of a university you want to search for:")
        university_name_input = input()
        # Convert the input to the desired format
        university_name_input = university_name_input.title()
        print("Please enter the criteria you want to sort by:")
        criteria_input = input()
        # Convert the input to the desired format
        criteria_input = criteria_input.lower().capitalize()
        print("Searching for university...")

        # Sort the universities
        sorted_universities = merge_sort(university_dict, criteria_input)

        # Search for a university
        result = hash_search(sorted_universities, university_name_input)

        # Assert that the search result is correct
        assert result is None or result['University name'] == university_name_input, "Search result is incorrect"

        # Print the result in the original format with overall score in percentage
        if result:
            rank = sorted_universities.index(result) + 1
            print(f"The rank and overall score of {university_name_input} is:")
            print(f"Rank: {rank}")
            print(f"University name: {result['University name']}")
            print(f"Entry standards: {result['Entry standards']}")
            print(f"Student satisfaction: {result['Student satisfaction']}")
            print(f"Research quality: {result['Research quality']}")
            print(f"Research intensity: {result['Research intensity']}")
            print(f"Graduate prospects: {result['Graduate prospects']}%")
            print(f"Overall score: {result['Overall score']}")
        else:
            print("University not found.")
    elif option == '3':
        break
    else:
        print("Invalid option. Please try again.")

```

```python
#Alternative
# Quick Sort
def quick_sort(univ_list, criteria):
    if len(univ_list) <= 1:
        return univ_list

    pivot_index = len(univ_list) // 2
    pivot = univ_list[pivot_index]

    # Split the list into two sub-arrays based on the pivot
    left = [univ for univ in univ_list if univ[criteria] > pivot[criteria]]
    middle = [univ for univ in univ_list if univ[criteria] == pivot[criteria]]
    right = [univ for univ in univ_list if univ[criteria] < pivot[criteria]]

    # Recursively sort the sub-arrays and combine the results
    return quick_sort(left, criteria) + middle + quick_sort(right, criteria)


# Linear Search
def linear_search(univ_list, target):
    for univ in univ_list:
        if univ['University name'] == target:
            return univ
    return None


# User Interaction for University Search
while True:
    print("Welcome to University Ranking")
    print("Please choose an option:")
    print("1. Display the ranking of all universities")
    print("2. Search for a specific university")
    print("3. Exit")
    option = input("Enter the selection between 1 or 2 or 3: ")

    # Assert that the user's input is a valid option
    assert option in ['1', '2', '3'], "Invalid option"

    if option == '1':
        print("Please enter the criteria you want to rank by:")
        criteria_input = input()
        # Convert the input to the desired format
        criteria_input = criteria_input.lower().capitalize()
        print("Ranking universities...")

        # Sort the universities based on the specified criteria
        sorted_universities = quick_sort(university_dict, criteria_input)

        execution_time = timeit.timeit(lambda: quick_sort(university_dict, criteria_input), number=1)

        # Print or store the execution time
        print(f"Execution time: {execution_time} seconds")

        # Print the ranking of all universities
        for i, univ in enumerate(sorted_universities):
            if criteria_input == 'Graduate prospects':
                print(f"Rank {i+1}: {univ['University name']} - {criteria_input}: {univ[criteria_input]}%")
            else:
                print(f"Rank {i+1}: {univ['University name']} - {criteria_input}: {univ[criteria_input]}")
    elif option == '2':
        print("Please enter the name of a university you want to search for:")
        university_name_input = input()
        # Convert the input to the desired format
        university_name_input = university_name_input.title()
        print("Please enter the criteria you want to rank by:")
        criteria_input = input()
        # Convert the input to the desired format
        criteria_input = criteria_input.lower().capitalize()
        print("Searching for university...")

        # Sort the universities based on the specified criteria
        sorted_universities = quick_sort(university_dict, criteria_input)

        # Search for the university
        result = linear_search(sorted_universities, university_name_input)

        if result:
            rank = sorted_universities.index(result) + 1
            print(f"The rank and overall score of {university_name_input} based on {criteria_input} is:")
            print(f"Rank: {rank}")
            print(f"University name: {result['University name']}")
            print(f"Entry standards: {result['Entry standards']}")
            print(f"Student satisfaction: {result['Student satisfaction']}")
            print(f"Research quality: {result['Research quality']}")
            print(f"Research intensity: {result['Research intensity']}")
            print(f"Graduate prospects: {result['Graduate prospects']}%")
            print(f"Overall score: {result['Overall score']}")
        else:
            print("University not found.")
    elif option == '3':
        break

```
